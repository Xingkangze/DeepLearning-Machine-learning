```python
import numpy as np
x = np.arange(1,4)
```

```python
x
```

```python
y = np.arange(10,15)
```

```python
y
```

```python
# 从坐标向量返回坐标矩阵。
# 它讲 x 变成了矩阵 z 的行向量，y 变成了矩阵 s 的列向量。
X,Y  = np.meshgrid(x,y)
```

```python
X
```

```python
Y
```

```python
X.ravel()
```

```python
Y.ravel()
```

```python
np.c_[X.ravel(),Y.ravel()]
```

```python
import math
Z =np.sin(X) + np.sin(Y);
Z
```

```python
#  根据x和y值绘制三维图形，其步骤就是，如果x和y只是代表取值范围的向量，
#     那么首先生成网格数据，
#     然后根据函数关系得到z值，使用plot3绘制曲面或者曲线。
import matplotlib.pyplot as  plt
from matplotlib.colors import ListedColormap
custom_cmap = ListedColormap(['#EF9A9A','#FFF59D','#90CAF9'])
plt.contourf(X,Y,Z,linewidth=5,cmap=custom_cmap)
plt.scatter(X[:,0],Y[:,0],color='red')
plt.scatter(X[:,1],Y[:,1],color='BLUE')
plt.scatter(X[:,2],Y[:,2],color='red')
plt.scatter(np.c_[X.ravel(),Y.ravel()][:,0],np.c_[X.ravel(),Y.ravel()][:,1],color='y')
plt.show()
```
