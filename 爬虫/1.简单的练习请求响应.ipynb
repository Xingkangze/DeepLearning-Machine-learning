```python
import requests
```

```python
response = requests.get('http://www.baidu.com')
```

```python
#获得网页的源代码
print(response.text)
```

```python
# 对网页的源代码进行分析
# 获取 响应头    字典形式的结构
response.headers
```

```python
# 获取状态码
response.status_code
```

```python
# 请求头
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/62.0.3202.62 Safari/537.36'}
```

```python
# 我们将请求头放进去我们的请求   重新发起请求
response = requests.get('http://www.baidu.com',headers=headers)
```

```python
response.status_code
```

```python
#  抓取一个图片
response1 = requests.get('https://static.zhihu.com/heifetz/logo.f6eef03354bbf46f4e04.png',headers=headers)
```

```python
# 获取到响应体的二进制格式
response1.content
```

```python
with open(r'F:\123\爬虫图片\\知乎图.jpg','wb')  as f:
    f.write(response1.content)
    f.close()
```
