总结一下python中常用的匹配符：
\w     可以匹配一个字母或者数字
\d     匹配数字
\d+    可以匹配多个数字
\d+?   可以匹配一部分数字（一组）
^      匹配行首
$      匹配行尾
^\d    第一个必须为数字
\d$    最后一个必须为数字
\s     匹配一个空格
\d{3,8}          匹配3-8个数字
[0-9a-zA-Z\_]    可以匹配一个数字、字母或者下划线；
[0-9a-zA-Z\_]+   可以匹配至少由一个数字、字母或者下划线组成的字符串，比如'a100'，'0_Z'，'Py3000'等等；
[a-zA-Z\_][0-9a-zA-Z\_]*   可以匹配由字母或下划线开头，后接任意个由一个数字、字母或者下划线组成的字符串，也就是Python合法的变量；
[a-zA-Z\_][0-9a-zA-Z\_]{0, 19}   更精确地限制了变量的长度是1-20个字符（前面1个字符+后面最多19个字符）。
.    匹配任意的字符
*    匹配任意个字符（包括0个）
？   匹配0个或1个字符，尽可能少的匹配
+    匹配至少一个字符
{n}      n个字符
{n,m}    n-m个字符

# re.match 
re.match 尝试从字符串的起始位置匹配一个模式，如果不是起始位置匹配成功的话，match()就返回none

re.match(pattern,string,flag=0)

```python
import re

content = 'hello 123 4567 World_This is a Regex Deamo'
print(len(content))
result = re.match('^hello\s\d{3}\s\d{4}\s\w{10}.*Deamo$',content)

print(result)
print(result.group())  # 返回匹配结果
print(result.span())
```

# 泛匹配

```python
import re 
content = 'hello 123 4567 World_This is a Regex Deamo'
result = re.match('^hello.*Deamo$',content)
print(result)
print(result.group())  # 返回匹配结果
print(result.span())
```

# 匹配目标

```python
import re 
content = 'hello 123 4567 World_This is a Regex Deamo'
result = re.match('^hello\s([\d\s]+)\s.* Deamo$',content)
print(result)
print(result.group())  # 返回匹配结果
print(result.group(1))
print(result.span())
```

# 贪婪匹配

```python
import re 
content = 'hello 123 4567 World_This is a Regex Deamo'
result = re.match('^he.*([\d]+).*mo$',content)
print(result)
print(result.group())  # 返回匹配结果
print(result.group(1))
print(result.span())
#我们本来想匹配1234657   结果只出来 7
```

# 非贪婪匹配

```python
import re 
content = 'hello 123 4567 World_This is a Regex Deamo'
# ? 指定为非贪婪  ？   尽可能少的字符
result = re.match('^he.*?([\d\s]+).*mo$',content)
print(result)
print(result.group())  # 返回匹配结果
print(result.group(1))
print(result.span())
```

# 匹配模式

```python

修饰符

描述

re.I

使匹配对大小写不敏感

re.L

做本地化识别（locale-aware）匹配

re.M

多行匹配，影响^和$

re.S

使.匹配包括换行在内的所有字符

re.U

根据Unicode字符集解析字符。这个标志影响\w、\W、 \b和\B

re.X

该标志通过给予你更灵活的格式以便你将正则表达式写得更易于理解
```

```python
import re
content = '''hello 123 4567 World_This is a Regex Deamo
is a Demamoasdas
'''
#本来的 .* 是不可以匹配回车的  我们加一个匹配模式   re.S  就可以匹配了
result = re.match('^he.*?([\d\s]+).*sdas$',content,re.S)
print(result)
print(result.group())  # 返回匹配结果
print(result.group(1))
print(result.span())
```

# 转义

```python
import re
content = 'hello 123 4567 World$This is a Regex Deamo'
result = re.match('^he.*?([\d\s]+)\w+\$.*Deamo$',content)
print(result)
print(result.group())  # 返回匹配结果
print(result.group(1))
print(result.span())
```

# re.search
re.search 扫描整个字符串并返回第一个成功的匹配

```python
# re.match 匹配是从第一个字符开始匹配 如果匹配不上  为none
import re
content = 'Ehello 123 4567 World$This is a Regex Deamo'
result = re.match('he.*?([\d\s]+)\w+\$.*Deamo$',content)
print(result)
```

```python
# re.search 是从全局匹配 
import re
content = 'Ehello 123 4567 World$This is a Regex Deamo'
result = re.search('he.*?([\d\s]+)\w+\$.*Deamo$',content)
print(result)
```

# re.findall
搜索字符串，以列表形式返回全部能匹配的子串

```python
import re
html = '''<div id="songs-list">
    <h2 class="title">经典老歌</h2>
    <p class="introduction">
        经典老歌列表
    </p>
    <ul id="list" class="list-group">
        <li data-view="2">一路上有你</li>
        <li data-view="7">
            <a href="/2.mp3" singer="任贤齐">沧海一声笑</a>
        </li>
        <li data-view="4" class="active">
            <a href="/3.mp3" singer="齐秦">往事随风</a>
        </li>
        <li data-view="6"><a href="/4.mp3" singer="beyond">光辉岁月</a></li>
        <li data-view="5"><a href="/5.mp3" singer="陈慧琳">记事本</a></li>
        <li data-view="5">
            <a href="/6.mp3" singer="邓丽君">但愿人长久</a>
        </li>
    </ul>
</div>'''

results = re.findall('<li.*?href="(.*?)".*?singer="(.*?)">(.*?)</a>',html,re.S)
# 本来是一个列表
if results:
    print(results)
for result in results:
    print(result)
```

# re.compile
将正则字符串编译为正则表达式对象
以便于复用我们的正则表达式

```python
import re
content = '''Ehello 123 4567 World$This is a 
Regex Deamo''';
pattern = re.compile('Ehell.*\$.*Deamo$',re.S)
result = re.match(pattern,content)
print(result)
```
