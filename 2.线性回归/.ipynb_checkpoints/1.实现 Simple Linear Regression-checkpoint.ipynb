```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,6,dtype=float)
x
```

```python
y = np.array([1.,3.,2.,3.,5.,])
y
```

```python
plt.scatter(x,y)
plt.axis([0,6,0,6])
plt.show()
```

## 来实现我们具体的公式

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
```

```python
num = 0.0    #分子
d = 0.0      #分母
for x_i,y_i in zip(x,y):
    num +=(x_i-x_mean)*(y_i-y_mean)
    d +=(x_i-x_mean) **2
```

```python
a = num/d
a
```

```python
b = y_mean-a*x_mean
b
```

```python
y_hat =a * x + b
```

```python
plt.scatter(x,y)
plt.plot(x,y_hat,color='r')
plt.axis([0,6,0,6])
plt.show()
```

```python
x_predict=6
y_predict = a*x_predict+b
```

```python
y_predict
```

```python
plt.scatter(x,y)
plt.plot(x,y_hat,color='r')
plt.scatter(x_predict,y_predict,color='g')
plt.axis([0,9,0,6])
plt.show()
```
