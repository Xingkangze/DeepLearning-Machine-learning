# matplotlib 练习
## Figure和Subplot


```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
mpl.rcParams['font.sans-serif'] = ['SimHei']
mpl.rcParams['axes.unicode_minus'] = False
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(212)
ax1.hist(np.random.rand(50).cumsum(),bins=20,normed=1,alpha=0.5,color='k')
ax2.scatter(np.random.rand(300),3*np.random.rand(300),color='b',marker='^',alpha=0.3)
ax3.plot(np.random.rand(50))
plt.show()
```

### subplots

```python
plt.figure()
plot,axs = plt.subplots(2,2,True,True)
plot.suptitle('subplots')
colors = ['k','y','b','r']
for i in range(2):
    for j in range(2):
        axs[i][j].hist(np.random.rand(300).cumsum(),bins=30,normed=1,alpha=0.5,color=colors[i+j])
plt.subplots_adjust(wspace=0,hspace=0)  #   间隔 
plt.show()
```

```python
plt.axis([0,5,0,20])
plt.title('My First Chart',fontsize=20,fontname='Times New Roman')  #  字体
plt.xlabel('Counting',color='gray')
plt.ylabel('square values',color='gray')
plt.text(1,1.5,'First')
plt.text(2,4.5,'scecond')
plt.text(3,9.5,'Third')
plt.text(4,16.5,'Forth')
plt.text(1,11.5,r'$y=x^2$',fontsize=20,bbox={'facecolor':'red','alpha':0.2})
plt.grid(True)  #  设置网格的开关
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.plot([1,2,3,4],[0.8,3.5,8,15],'g^')
plt.plot([1,2,3,4],[0.5,2.5,7,14],'b*')
plt.legend(['first1','second2'],loc=4)  # loc 表示位置
plt.show()
```

```python
x = np.arange(-5,5)
y = np.sin(x)
plt.axis([-5,5,-5,5])
plt.plot(x,y,color='g',linestyle='dashed',marker='o')
plt.text(-3,3,r'$y=sin(x)$',fontsize=20,bbox={'facecolor':'yellow','alpha':0.2})
plt.show()
```
