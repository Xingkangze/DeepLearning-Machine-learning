## Matplotlib：Subplot 子图
- Subplot：子图，figure对象下创建一个或多个subplot对象(即axes)用于绘制图像。
- subplot(numRows, numCols, plotNum)

````python
x = [1,2,3]
y = [5,7,4]
x1= [1,2,3]
y1= [10,14,12]

plt.figure(figsize=(8,8))
plt.subplot(221)   #   2行  2列   第几个
plt.plot(x,y,'r',marker='^',alpha=0.5,markersize=10,label='1') 
plt.xticks([1,2,3,4,5,6],[str(i)+'月' for i in range(1,7)])
# plt.xticks([1,2,3,4,5,6])
plt.yticks(np.arange(0,16,3),['2012年','2013年','2014年','2015年','2016年'])

plt.legend()
plt.subplot(224)
plt.plot(x1,y1,'b',marker='>',alpha=0.5,markersize=10,label='2')
plt.xticks([1,2,3,4,5,6],[str(i)+'月' for i in range(1,7)])
# plt.xticks([1,2,3,4,5,6])
plt.yticks(np.arange(0,16,3),['2012年','2013年','2014年','2015年','2016年'])

plt.legend()
plt.title('销售数据')
plt.show()
````

## 通过面向对象的形式  来画出子图
```
面向对象的形式：
fig = plt.figure()
    Figure实例
    可以添加Axes实例
ax = fig.add_subplot(111)
    返回Axes实例
    参数1，子图的总行数
    参数2，子图的总列数
    参数3，子图位置
在Figure上添加子图的常用方法
```

````python
np.random.randn(50).cumsum()    #   累计求和
````

````python
fig = plt.figure()
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(212)

ax1.plot(np.random.randn(50),'r--',marker='^')
ax2.plot(np.random.randn(50),'b--',marker='^')
ax3.plot(np.random.randn(50),'y--',marker='^')

plt.show()
````

## Subplots():返回一个图像和多个子图
```
参数：  nrows=x, ncols=x, sharex=True, sharey=False, 
        	gridspec_kw={'height_ratios':[2,2,1,1]}
例：
fig, ax = plt.subplots(2,2),其中参数分别代表子图的行数和列数，一共有 2x2 个图像。函数返回一个figure图像和一个子图ax的array列表。
```

````python
plt.figure()
plot,axis = plt.subplots(2,2)
plot.suptitle('test')
axis[0][0].plot(np.arange(10),'o--')
axis[0][1].plot(np.arange(10),'o--')
axis[1][0].plot(np.arange(10),'o--')
axis[1][1].plot(np.arange(10),'o--')
plt.show()
````

````python
plt.figure()
plot1,axes = plt.subplots(ncols=4,nrows=1,sharex=True,sharey=False)  #   sharex  共享Y轴
plot1.suptitle('哈哈哈')
axes[0].plot(np.arange(10),'o--')
axes[0].set_title('Sharing Y axis')
axes[1].plot(np.arange(10),'o--')
axes[2].plot(np.arange(10),'o--')
axes[3].plot(np.arange(10),'o--')
plt.show()
````

参数： x:数据标量 height：高 width:款 bottom：底端对应Y轴align:对齐如果为 "居中", 则将x参数解释为条形中心的坐标。如果 "边缘", 将条形按其左边缘对齐要对齐右边缘的条形图, 可传递负的宽度和对align='edge'

```
用柱状图表示第一季度的进出口数据
进口数据：
    1月份  600亿美元
    2月份  700亿美元
    3月份  800亿美元
出口数据：
    1月份  500亿美元
    2月份  900亿美元
    3月份  1000亿美元
```

````python
plt.figure()
x = np.arange(1,4)
y1 = np.array([600,700,800])
y2=np.array([500.0,900.0,1000.0])
#facecolor柱状图里填充的颜色
#edgecolor是边框的颜色
# plt.bar(x, y2, facecolor='white', edgecolor='#000000')
plt.bar(x,y1,facecolor='red',width = 0.35)
plt.bar(x+0.35,y2,color='b',width = 0.35)
for x11,y11 in zip(x,y1):
    plt.text(x11, y11, '{}'.format(y11), ha='center', va= 'bottom')
for x21,y21 in zip(x,y2):
    plt.text(x21+0.35,y21,'{}'.format(y21),ha='center',va='bottom')
plt.xlabel('月份')
plt.ylabel('销售额')
plt.xlim(0,4)
plt.ylim(0,1000)
plt.title('第一季度进口数据')


plt.xticks([1.5,2.5,3.5],[str(i)+'月' for i in range(1,7)])
plt.yticks([0,200,400,600,800,1000.0])


plt.show()
````

## Matplotlib直方图
```
matplotlib.pyplot.hist(
	x,bins=10,range=None,normed=False,weights=None,cumulative=False,bottom=None,
	histtype='bar', align='mid',orientation=u'vertical', rwidth=None, log=False,color=None, 	label=None,stacked=False,hold=None,**kwargs) 
– x: 一个列表或者多个列表（表示不同数据集，长度可以不一致）
– range: 元组
– weights: x里每个元素对bin高度的贡献（默认为1）
– bottom: 数字或者长度为bins的列表
– histtype: ['bar' | 'barstacked' | 'step' | 'stepfilled'] 
– align: ['left' | 'mid' | 'right'] 
– orientation: ['horizontal' | 'vertical'] 
– rwidth: bar相对bin的宽度
– color: 一种颜色或者颜色列表（针对不同数据集）
```

````python
x =np.random.randn(1000)
# print(x)
plt.hist(x,100,normed=1,alpha=0.5,color='red')
plt.text(-3,0.4,r'$\mu=0,\ \sigma=1$')
plt.show()
````

## 绘制散点图
```
matplotlib.pyplot. scatter (x, y, s=none, c=none, marker=none, cmap= None, 
                            norm=none, vmin=none, vmax=none, alpha=none, Linewidths=none,
                            verts=none, edgecolors=none, hold=none, data= None, **kwargs)
参数：
    x,y:相同长度的数组序列
    s  :散点的大小标量或形同数组，可选参数，默认20
    c  :散点的色彩或颜色序列，可选
    maker:标记风格，可选，默认是‘o’
    norm:数据的亮度
    alpha:散点的透明度
```

````python
x = [1,2,3,4,5,6,7,8]
y = [5,2,4,2,1,4,5,2]
T = np.random.rand(8)*125
plt.scatter(x,y,c=T,s=25,marker='^',alpha=0.5)
plt.show()
````
