````python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
````

````python
x = np.linspace(0,10,100)
x
````

````python
y = np.sin(x)
y
````

````python
plt.plot(x,y)
plt.show()
````

````python
cosy = np.cos(x)
cosy
````

````python
siny = y.copy()
````

````python
plt.plot(x,siny)
plt.plot(x,cosy,color='red')
plt.show()
````

````python
plt.plot(x,siny,linestyle='-.')
plt.plot(x,cosy,color='red',linestyle='--')      #
plt.show()
````

````python
plt.plot(x,siny,linestyle='-.')
plt.plot(x,cosy,color='red',linestyle='--')      #
plt.xlim(-5,15)           #x
plt.ylim(0,1.5)           #y
plt.show()
````

````python
plt.plot(x,siny,color='blue')
plt.plot(x,cosy,color='red',linestyle='--')
plt.axis([-1,11,-2,2])          #  x     y
plt.show()
````

````python
plt.plot(x,siny)
plt.plot(x,cosy)
plt.axis([-1,11,-2,2])
plt.xlabel('x axis')               #X
plt.ylabel('y axis')               #y
plt.show()
````

````python
plt.plot(x,siny,label='sin(x)')    #x
plt.plot(x,cosy,label='cos(x)')    #y
plt.axis([-1,11,-2,2])
plt.xlabel('x axis')               #X
plt.ylabel('y axis')               #y
plt.legend()               #
plt.title('welcome to the ML world')     # 
plt.show()
````

## Subplots():
```
  nrows=x, ncols=x, sharex=True, sharey=False, 
        	gridspec_kw={'height_ratios':[2,2,1,1]}

fig, ax = plt.subplots(2,2), 2x2 figureaxarray
```

````python
plt.figure()
plot,axis = plt.subplots(2,2)
plot.suptitle('test')
axis[0][0].plot(np.arange(10),'o--')
axis[0][1].plot(np.arange(10),'o--')
axis[1][0].plot(np.arange(10),'o--')
axis[1][1].plot(np.arange(10),'o--')
plt.show()
````

````python
plt.figure()
plot1,axes = plt.subplots(ncols=4,nrows=1,sharex=True,sharey=False)  #   sharex  Y
plot1.suptitle('')
axes[0].plot(np.arange(10),'o--')
axes[0].set_title('Sharing Y axis')
axes[1].plot(np.arange(10),'o--')
axes[2].plot(np.arange(10),'o--')
axes[3].plot(np.arange(10),'o--')
plt.show()
````

````python
plt.figure()
x = np.arange(1,4)
y1 = np.array([600,700,800])
y2=np.array([500.0,900.0,1000.0])
#facecolor
#edgecolor
# plt.bar(x, y2, facecolor='white', edgecolor='#000000')
plt.bar(x,y1,facecolor='red',width = 0.35)
plt.bar(x+0.35,y2,color='b',width = 0.35)
for x11,y11 in zip(x,y1):
    plt.text(x11, y11, '{}'.format(y11), ha='center', va= 'bottom')
for x21,y21 in zip(x,y2):
    plt.text(x21+0.35,y21,'{}'.format(y21),ha='center',va='bottom')
plt.xlabel('')
plt.ylabel('')
plt.xlim(0,4)
plt.ylim(0,1000)
plt.title('')


plt.xticks([1.5,2.5,3.5],[str(i)+'' for i in range(1,7)])
plt.yticks([0,200,400,600,800,1000.0])


plt.show()
````
