```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
# 构建数据
points_num = 100
vectors = []
# 生成正态分布100个点 y = 0.1*x+0.2 
for i in range(points_num):
    x1 = np.random.normal(0.0,0.66)
    y1 = 0.1*x1+0.2+np.random.normal(0.0,0.04)
    vectors.append([x1,y1])
```

```python
x_data = [v[0] for v in vectors ]
y_data = [v[1] for v in vectors ]
plt.plot(x_data,y_data,'r*',label='original data')
plt.title('Linear Regression')
plt.show()
```

```python
x = np.random.normal(0.0,0.66,size=100)
X = x.reshape(-1,1)
y = 0.1*X+0.2+np.random.normal(0.0,0.04,size=(len(x))).reshape(-1,1)
#vector2 =np.concatenate([X,y],axis=1) 
```

```python
plt.plot(X,y,'r*',label='original data')
plt.title('Linear Regression')
plt.show()
```

```python
# 构建线性回归模型
```

```python
w = tf.Variable(tf.random_uniform([1],-1.0,1.0)) #初始化权重
b = tf.Variable(tf.zeros([1])) # 偏差初始化为0
y_train = w *X +b # 模型计算出来的 y

# 定义损失函数 loss function
# 对 tensor所有维度计算（y-y_data）^2和/N
loss = tf.reduce_mean(tf.square(y_train-y))
# 用梯度下降的优化器优化 loss function()
optimizer = tf.train.GradientDescentOptimizer(0.5)# 0.5 学习率
train = optimizer.minimize(loss)
```

```python
# 初始化数据流图中的所有变量

init = tf.global_variables_initializer()


# 创建会话
with tf.Session() as sess:
    sess.run(init)

    # 训练
    for step in range(20):
        sess.run(train)
        #打印出每一步的损失权重 和偏差
        print('step=%d,loss=%f,[weight=%f Bias =%f]'%(step,sess.run(loss),
                                                      sess.run(w),sess.run(b)))

    # 绘制图像
    plt.plot(X,y,'r*',label='original data')
    plt.title('Linear Regression')
    plt.plot(X,sess.run(w)*X+sess.run(b),label='fitted line') # 拟合的直线
    plt.legend()
    plt.show()
```
