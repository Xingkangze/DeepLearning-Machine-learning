```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

x = np.random.normal(0.0,0.55,size=500)
X = x.reshape(-1,1)
Y = 0.1*X+0.2+np.random.normal(0.0,0.04,size=(len(X))).reshape(-1,1)
```

```python
plt.scatter(X,Y,c='r')
plt.show()
```

```python
# 构造模型
# 权重初始化 w
W = tf.Variable(tf.random_uniform([1],-1.0,1.0,dtype=tf.float32),name='W')
# 偏差 b
b = tf.Variable(tf.zeros([1]),name='b')
# 经过计算得出的预测值 y
y = W*X+b
# 以预估值y 和实际值y之间的均方误差作为损失
loss = tf.reduce_mean(tf.square(y-Y),name='loss')
# 采用梯度下降法  学习率
optimizer = tf.train.GradientDescentOptimizer(0.5)
# 训练的过程就是最小化误差值
train = optimizer.minimize(loss,name='train')
# 初始化
init = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
    for i in range(50):
        sess.run(train)
        print('w= ',sess.run(W),'b=',sess.run(b),'loss=',sess.run(loss))
    plt.scatter(X,Y,c='r')
    plt.plot(X,sess.run(W)*X+sess.run(b))
    plt.show()   
```
