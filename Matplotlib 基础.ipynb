```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
cosy = np.cos(x)
cosy
```

```python
siny = y.copy()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color='red')
plt.show()
```

```python
plt.plot(x,siny,linestyle='-.')
plt.plot(x,cosy,color='red',linestyle='--')      #
plt.show()
```

```python
plt.plot(x,siny,linestyle='-.')
plt.plot(x,cosy,color='red',linestyle='--')      #
plt.xlim(-5,15)           #x
plt.ylim(0,1.5)           #y
plt.show()
```

```python
plt.plot(x,siny,color='blue')
plt.plot(x,cosy,color='red',linestyle='--')
plt.axis([-1,11,-2,2])          #  x     y
plt.show()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy)
plt.axis([-1,11,-2,2])
plt.xlabel('x axis')               #X
plt.ylabel('y axis')               #y
plt.show()
```

```python
plt.plot(x,siny,label='sin(x)')    #x
plt.plot(x,cosy,label='cos(x)')    #y
plt.axis([-1,11,-2,2])
plt.xlabel('x axis')               #X
plt.ylabel('y axis')               #y
plt.legend()               #
plt.title('welcome to the ML world')     # 
plt.show()
```

## Scatter Plot     

```python
plt.scatter(x,siny,color='red')
plt.show()
```

```python
plt.scatter(x,siny,color='red',label='sin(x)')
plt.scatter(x,cosy,color='blue',label='cos(x)')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('kangze')
plt.show()
```

```python
x = np.random.normal(0,1,10000)
y = np.random.normal(0,1,10000)
plt.scatter(x,y,color='red',label='sandian',alpha=0.1)   #alpha  0-1 
plt.title('z t t')
plt.legend()
plt.show()
```
