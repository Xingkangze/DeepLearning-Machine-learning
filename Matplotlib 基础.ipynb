```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
cosy = np.cos(x)
cosy
```

```python
siny = y.copy()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color='red')
plt.show()
```

```python
plt.plot(x,siny,linestyle='-.')
plt.plot(x,cosy,color='red',linestyle='--')      #线条类型
plt.show()
```

```python
plt.plot(x,siny,linestyle='-.')
plt.plot(x,cosy,color='red',linestyle='--')      #线条类型
plt.xlim(-5,15)           #x的区间范围
plt.ylim(0,1.5)           #y的区间范围
plt.show()
```

```python
plt.plot(x,siny,color='blue')
plt.plot(x,cosy,color='red',linestyle='--')
plt.axis([-1,11,-2,2])          #同时调节  x轴   和  y
plt.show()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy)
plt.axis([-1,11,-2,2])
plt.xlabel('x axis')               #X轴写上一个名字
plt.ylabel('y axis')               #y轴写上一个名字
plt.show()
```

```python
plt.plot(x,siny,label='sin(x)')    #在图中标注x的标签
plt.plot(x,cosy,label='cos(x)')    #在图中标注y的标签
plt.axis([-1,11,-2,2])
plt.xlabel('x axis')               #X轴写上一个名字
plt.ylabel('y axis')               #y轴写上一个名字
plt.legend()               #在图中显示出来
plt.title('welcome to the ML world')     #图的名字 
plt.show()
```

## Scatter Plot     散点图

```python
plt.scatter(x,siny,color='red')
plt.show()
```

```python
plt.scatter(x,siny,color='red',label='sin(x)')
plt.scatter(x,cosy,color='blue',label='cos(x)')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('kangze')
plt.show()
```

```python
x = np.random.normal(0,1,10000)
y = np.random.normal(0,1,10000)
plt.scatter(x,y,color='red',label='sandian',alpha=0.1)   #alpha  0-1 代表透明度
plt.title('z t t')
plt.legend()
plt.show()
```
