## ROC曲线

```python
# 分类准确度几个指标
def TN(y_true, y_predict):
    '''这个是正确的预测了值为0的数'''
    assert len(y_true) == len(y_predict)
    return np.sum((y_true == 0) & (y_predict == 0))


def FP(y_true, y_predict):
    '''预测错误   本来正确为0  我们将其 预测为1'''
    assert len(y_true) == len(y_predict)
    return np.sum((y_true == 0) & (y_predict == 1))


def FN(y_true, y_predict):
    '''真实值为1 但是我们预测为0'''
    assert len(y_true) == len(y_predict)
    return np.sum((y_true == 1) & (y_predict == 0))


def TP(y_true, y_predict):
    '''真实值为1 预测值为1'''
    assert len(y_true) == len(y_predict)
    return np.sum((y_true == 1) & (y_predict == 1))
```

```python
# TPR
def TPR(y_true, y_predict):
    assert len(y_true) == len(y_predict)
    tp = TP(y_true, y_predict)
    fn = FN(y_true, y_predict)
    try:
        return tp / (tp + fn)
    except:
        return 0.0


# FPR
def FPR(y_true, y_predict):
    assert len(y_true) == len(y_predict)
    fp = FP(y_true, y_predict)
    tn = TN(y_true, y_predict)
    try:
        return fp / (fp + tn)
    except:
        return 0.0
```

```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
digits = datasets.load_digits()
X = digits.data
y = digits.target.copy()

y[digits.target==9] = 1
y[digits.target!=9] =0
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=666)
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(X_train,y_train)
#决策的分数值
decision_score = log_reg.decision_function(X_test)
```

```python
thresholds = np.arange(np.min(decision_score),np.max(decision_score),0.1)
%run d:/Python1/Muke_MLB/Matrix.py
fprs = []
tprs = []
for threshold in thresholds:
    y_predict = np.array(decision_score>=threshold,dtype='int')
    fprs.append(FPR(y_test,y_predict))
    tprs.append(TPR(y_test,y_predict))
```

```python
'''绘制roc曲线'''
plt.plot(fprs,tprs)
plt.show()
'''这个曲线很明显随着fpr的逐渐增高，我们的tpr也逐渐变大'''
```

## scikit-learn 中的ROC

```python
from sklearn.metrics import roc_curve

fprs , tprs , thresholds =roc_curve(y_test,decision_score)
```

```python
plt.plot(fprs,tprs)
plt.show()
```

```python
'''对于ROC曲线来说，我们通常关注的是曲线下面的面积大小，曲线下面的面积越大
   代表我们的分类效果越好，这是因为当我们的X越小的时候，fpr值越小，我们犯得fp
   的错误越小的时候，他的tpr越大，那么整体也就会越高，面积也就会越大 ，所以ROC
   的曲线的面积可以作为分类算法的指标
'''
```

```python
'''求面积    auc代表曲线下面的面积 面积最大值为1'''
from sklearn.metrics import roc_auc_score
'这个求法是先求出曲线  curve，然后求auc'
roc_auc_score(y_test,decision_score)
```
