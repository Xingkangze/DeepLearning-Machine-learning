# Lasso

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
x = np.random.uniform(-3,3,size=100)
X = x.reshape(-1,1)
y = 0.5*x + 3 +np.random.normal(0,1,size=100)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=666)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
```

```python
def Polynomia_Regression(degree):
    return Pipeline(
    [
        ('poly',PolynomialFeatures(degree=degree)),
        ('sta_sca',StandardScaler()),
        ('Lin',LinearRegression())
    ]
    )
```

```python
from sklearn.metrics import  mean_squared_error

poly = Polynomia_Regression(20)
poly.fit(X_train,y_train)
y_pridect = poly.predict(X_test)
mean_squared_error(y_test,y_pridect)
```

```python
def plot_modle(plot1):
    X_plot = np.linspace(-3,3,100).reshape(-1,1)
    y_plot = plot1.predict(X_plot)
    plt.scatter(X,y)
    plt.plot(X_plot,y_plot,color='r')
    plt.axis([-3,3,0,7])
    plt.show()
```

```python
 plot_modle(poly)
```

# Lasso

```python
from sklearn.linear_model import Lasso
```

```python
# 传入的是一个Lasso  正则化
def Lasso_Regression(degree,alpha):
    return Pipeline(
    [
        ('poly',PolynomialFeatures(degree=degree)),
        ('sta_sca',StandardScaler()),
        ('Lasso',Lasso(alpha=alpha))
    ]
    )
```

```python
#下面都是在选取他的正则化的alpha的值
lasso_reg = Lasso_Regression(20,0.1)
lasso_reg.fit(X_train,y_train)
y1_predict = lasso_reg.predict(X_test)
mean_squared_error(y1_predict,y_test)
```

```python
plot_modle(lasso_reg)
```

```python
lasso1_reg = Lasso_Regression(20,10)
lasso1_reg.fit(X_train,y_train)
y2_predict = lasso1_reg.predict(X_test)
mean_squared_error(y2_predict,y_test)
```

```python
plot_modle(lasso1_reg)
```

```python
lasso2_reg = Lasso_Regression(20,1)
lasso2_reg.fit(X_train,y_train)
y3_predict = lasso2_reg.predict(X_test)
mean_squared_error(y3_predict,y_test)
```

```python
plot_modle(lasso2_reg)
```

```python
lasso2_reg = Lasso_Regression(20,0.5)
lasso2_reg.fit(X_train,y_train)
y3_predict = lasso2_reg.predict(X_test)
mean_squared_error(y3_predict,y_test)
```

```python
plot_modle(lasso2_reg)
```
